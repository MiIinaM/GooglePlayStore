# Data Converting

```python
googleplaystore_data.columns = [c.replace(' ','_') for c in googleplaystore_data.columns]
```

```python
googleplaystore_data.dropna(subset=['Type'], inplace=True)
googleplaystore_data.dropna(subset=['Content_Rating'], inplace=True)
googleplaystore_data.dropna(subset=['Current_Ver'], inplace=True)
googleplaystore_data.dropna(subset=['Android_Ver'], inplace=True)
```

```python
googleplaystore_data.Reviews = googleplaystore_data.Reviews.astype(float)
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

googleplaystore_data.Size = googleplaystore_data.Size.apply(value_to_float)
```

```python
googleplaystore_data.Installs = [i.replace('+', '') for i in googleplaystore_data.Installs]
```

```python
googleplaystore_data.Installs
```

```python
googleplaystore_data.Installs = [i.replace(',', '') for i in googleplaystore_data.Installs]
```

```python
googleplaystore_data.Installs = googleplaystore_data.Installs.astype(float)
```

```python
googleplaystore_data.Price = [j.replace('$', '') for j in googleplaystore_data.Price]
```

```python
googleplaystore_data.Price = googleplaystore_data.Price.astype(float)
```

```python
# look that the info of the dataframe after converting
googleplaystore_data.info()
```

# Missing Data

```python
# check on null values
googleplaystore_data.isna().sum()
```

```python
def replace_with_median(series):
    """
    Given a series, replace the rows with null values 
    with median values
    """
    return series.fillna(series.median())

googleplaystore_data['Rating'] = googleplaystore_data['Rating'].transform(replace_with_median)
googleplaystore_data['Rating'] = googleplaystore_data['Rating'].astype(float)
```

```python
# remove the record where 'Type' is having null value
index = googleplaystore_data[googleplaystore_data['Type'].isna()].index
googleplaystore_data.drop(axis=0, inplace=True, index=index)
```

```python
# check on null values
googleplaystore_data.isna().sum()
```

## Exploratory Analysis and Visualization

```python
# check on statistical information of the dataframe
googleplaystore_data.describe()
```

```python
# get the number of apps for each category
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
sns.countplot(x='Category', data=googleplaystore_data)
plt.title('Number of Apps Per Category')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

```python
# get the number of installs for each category
categories = googleplaystore_data.groupby('Category')
category_installs_sum_df = categories[['Installs']].sum()
category_installs_sum_df = category_installs_sum_df.reset_index()  # to convert groupby object into dataframe

plt.figure(figsize=(20, 5))
sns.barplot(x='Category', y='Installs', data=category_installs_sum_df)
plt.xticks(rotation=90)
plt.ylabel('Installs (e+10)')
plt.title('Number of Installs For Each Category')
plt.show()
```

```python
# show the distribution of rating
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=googleplaystore_data)
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

```python
# plot the graphs of reviews, size, installs and price per rating
rating_df = googleplaystore_data.groupby('Rating').sum().reset_index()

fig, axes = plt.subplots(1, 4, figsize=(14, 4))

axes[0].plot(rating_df['Rating'], rating_df['Reviews'], 'r')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Reviews')
axes[0].set_title('Reviews Per Rating')

axes[1].plot(rating_df['Rating'], rating_df['Size'], 'g')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Size')
axes[1].set_title('Size Per Rating')

axes[2].plot(rating_df['Rating'], rating_df['Installs'], 'g')
axes[2].set_xlabel('Rating')
axes[2].set_ylabel('Installs (e+10)')
axes[2].set_title('Installs Per Rating')

axes[3].plot(rating_df['Rating'], rating_df['Price'], 'k')
axes[3].set_xlabel('Rating')
axes[3].set_ylabel('Price')
axes[3].set_title('Price Per Rating')

plt.tight_layout(pad=2)
plt.show()
```

```python
# application type distribution
plt.figure(figsize=(10, 5))
sns.countplot(googleplaystore_data['Type'])
plt.title('Type Distribution')
plt.ylabel('Number of Apps')
plt.show()
```

# Asking and Answering Questions

```python
# 1. What is the top 5 apps on the basis of installs?
df = googleplaystore_data.sort_values(by=['Installs'], ascending=False)
df.head(5)
```

```python
# 2. What is the top 5 reviewed apps?
df = googleplaystore_data.groupby(by=['App', 'Category', 'Rating'])[['Reviews']].sum().reset_index()
df = df.sort_values(by=['Reviews'], ascending=False)
df.head(5)
```

```python
# 3. What is the top 5 expensive apps?
df = googleplaystore_data.sort_values(by=['Price'], ascending=False)
df.head(5)
```

```python
# 4. What is the top 3 most installed apps in Game category?
df = googleplaystore_data[googleplaystore_data['Category'] == 'GAME']
df = df.sort_values(by=['Installs'], ascending=False)
df.head(3)
```

```python
# 5. Which 5 apps from the 'FAMILY' category are having the lowest rating?
df = googleplaystore_data[googleplaystore_data['Category'] == 'FAMILY']
df = df.sort_values(by=['Rating'], ascending=True)
df.head(5)
```
